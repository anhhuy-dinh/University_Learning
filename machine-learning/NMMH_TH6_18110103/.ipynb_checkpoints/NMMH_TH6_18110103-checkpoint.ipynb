```python
data = pd.read_csv('https://raw.githubusercontent.com/huynhthanh98/ML/master/lab-06/bt1.csv')
data.head()
```

```python
X = data.iloc[:,[3, 4]].values

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(X[:, 0], X[:, 1])
ax.set_xlabel(data.columns[3])
ax.set_ylabel(data.columns[4])

plt.show()
```

```python
def plot_dendrogram(data, method='ward', decision_line=None, figsize=(10,5), leaf_rotation=90, leaf_font_size=5):
  import scipy.cluster.hierarchy as sch
  plt.figure(figsize=figsize)
  dendrogram = sch.dendrogram(sch.linkage(data, method=method), leaf_rotation=leaf_rotation, leaf_font_size=leaf_font_size)
  if decision_line is not None:  
    plt.axhline(y=decision_line, color='r', linestyle='--')
  plt.title('Dendrogram (%s method)' % method)
  plt.xlabel('Customers')
  plt.ylabel('Euclidean distances')
  plt.show()
```

```python
def plot_clustering(X, labels, ncol_legend=2, figsize=(10,5), xlabel=None, ylabel=None, legend=False):
  fig, ax = plt.subplots(figsize=figsize)
  label = np.unique(labels)
  for i in label:
      ax.scatter(X[labels == i][:, 0], X[labels == i][:, 1], marker='o', color=plt.cm.nipy_spectral(label[i] / len(label)), label = 'cluster_' + str(i))
  if legend:
    plt.legend(ncol=ncol_legend)
  if xlabel is not None:
    ax.set_xlabel(xlabel)
  if ylabel is not None:
    ax.set_ylabel(ylabel)
  plt.show()
```

**Ward Linkage**

```python
X_ward = X.copy()
plot_dendrogram(X_ward, method='ward', decision_line=150)
```

Dựa vào dendrogram trên, ta chọn số cụm là $5$.

```python
clustering = AgglomerativeClustering(linkage='ward', n_clusters=5)
clustering.fit(X_ward)
plot_clustering(X_ward, clustering.labels_, xlabel=data.columns[3], ylabel=data.columns[4])
```

**Complete Linkage**

```python
X_complete = X.copy()
linkage = 'complete'
plot_dendrogram(X_complete, method=linkage, decision_line=70)
```
