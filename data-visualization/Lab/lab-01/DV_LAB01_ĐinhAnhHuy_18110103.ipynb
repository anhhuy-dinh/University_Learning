```python
import numpy as np
import pandas as pd
```

### 1. Load data và xuất 5 bản ghi đầu tiên của data

```python
df = pd.read_csv('countries of the world.csv')
df.head()
```

### 2. Lấy dữ liệu từ các cột
Lấy dữ liệu từ các cột sau **'Country', 'Region', 'Population', 'Area (sq. mi.)', 'Coastline (coast/area ratio)', 'GDP ( per capita)'** (các yêu cầu dưới sẽ thực hiện trên dữ liệu này).

```python
cols = ['Country', 'Region', 'Population', 'Area (sq. mi.)', 'Coastline (coast/area ratio)', 'GDP ($ per capita)']
countries_df = df[cols]
countries_df.head()
```

### 3. Đổi tên các columns
Đổi tên các columns sau: **'Area (sq. mi.)' : 'Area' , 'Coastline (coast/area ratio)' : 'Coastline', 'GDP ($ per capita)' : 'GDP'**

```python
countries_df = countries_df.rename(columns={'Area (sq. mi.)' : 'Area' , 'Coastline (coast/area ratio)' : 'Coastline', 'GDP ($ per capita)' : 'GDP'})
```

### 15. Tên dư 1 dấu cách. Yêu cầu xoá dấu cách trong tên của columns Country
Dùng **apply** kết hợp **replace**

```python
countries_df['Country'] = countries_df['Country'].apply(lambda x: x.replace(" ", ""))
countries_df.head()
```

### Bonus
Ta thấy dtype của columns Coastline là object, nhưng khi nhìn vào giá trị thì ta sẽ nghĩ tới là numerical.
Yêu cầu: 
* Chuyển đổi dtype của cột **Coastline** sang **numerical(float)**, chú ý dấu phẩy ','.
* Sau đó, cho biết các nước (**Country**) không tiếp giáp với biển (**coastline = 0**).

```python
print(countries_df['Coastline'].dtypes)
countries_df['Coastline'].head()
```

```python
countries_df['Coastline'] = countries_df['Coastline'].apply(lambda x: x.replace(",", "."))
countries_df['Coastline'].head()
```

```python
countries_df['Coastline'] = countries_df['Coastline'].apply(pd.to_numeric, errors='coerce')
countries_df['Coastline'].head()
```
