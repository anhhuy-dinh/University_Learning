```python
# import basic libraries 
import numpy as np 
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
# import plot libraries 
import seaborn as sns 
import matplotlib.pyplot as plt 

#Mỗi dấu cách là một lần chạy mã

path='/content/CustomerChurn.csv'
df_customer=pd.read_csv(path)
df_customer.head(10)

path='/content/BigMartSales.csv'
df_mart=pd.read_csv(path)
df_mart.head(10)

#Consider dataset BigMart Sales
print('Columns s name of BigMart Sales dataset: \n',df_mart.columns)
print('Columns s name of Customer Churn dataset: \n',df_customer.columns)
print('Shape of BigMart Sales dataset before drop null values: ',df_mart.shape)
print('Shape of Customer Churn dataset before drop null values: ',df_customer.shape)
df_mart=df_mart.dropna()
df_customer=df_customer.dropna()
print('Shape of BigMart Sales dataset after drop null values: ',df_mart.shape)
print('Shape of Customer Churn dataset afterr drop null values: ',df_customer.shape)

-Phần Văn Bản
/*
Với mỗi tiêu chí/ thuộc tính của dữ liệu CustomerChurn hay BigMartSales chọn một hình vẽ EDA phù hợp kèm theo
nhận xét của bạn về tiêu chí/thuộc tính đó:
BigMart Sales
*/

print('Type of each features of BigMart Sales: \n',df_mart.dtypes)


sns.catplot(x='Item_Fat_Content',kind='count',data=df_mart)
#Ta thấy, Item fat content có số lượng Low Fat là chiếm nhiều nhất và nhỏ nhất là low fat.

sns.catplot(x='Item_Type',kind='count',data=df_mart).set_xticklabels(rotation=90)
#Item type chiếm số lượng lớn nhất là Snack Foods và Fruits and Vegetables

f = plt.figure(figsize=(8,8))
gs = f.add_gridspec(2, 3)
with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 0])   
with sns.axes_style("white"):
    ax = f.add_subplot(gs[0, 1])
    sns.distplot(df_mart.Item_Visibility)
with sns.axes_style("ticks"):
    ax = f.add_subplot(gs[0, 2])
    sns.distplot(df_mart.Item_MRP,bins=10)
with sns.axes_style("white"):
    sns.distplot(df_mart.Item_Weight,bins=20)
  ax = f.add_subplot(gs[0, 0])
  sns.displot(df_mart.Item_Outlet_Sales,bins=20)
#Theo distplot thì Item_weight trông không tuân theo phân phối chuẩn và tập trung chủ yếu là mức từ 5 tới 10.
#Ta thấy item_visibility tuân theo phân phối chuẩn nhưng hơi lệch phải và tập trung chủ yếu ở mức từ 0.03 tới 0.05
#Item_MRP tập trung nhiều ở 75-100 và từ 150 - 200
#Item_Outlet_sales tuân theo phân phối chuẩn và hơi lệch phải với đỉnh ở 1500 - 2000

sns.catplot(x='Outlet_Size',kind='count',data=df_mart)
sns.catplot(x='Outlet_Location_Type',kind='count',data=df_mart)
#Số lượng outlet_size của Medium và Small khá đều nhau và tương tự cho Tier 1, tier 2 cho outlet_location_type

#Outlet_Establishment_Year
sns.catplot(x='Outlet_Establishment_Year',kind='count',data=df_mart)
#Năm establish outlet là đều nhau với các năm là 1987, 1997, 1999, 2004, 2009

print('Type of each features of Customer Churn: \n',df_customer.dtypes)

f = plt.figure(figsize=(8,8))
gs = f.add_gridspec(2, 3)
ax = f.add_subplot(gs[0, 0])
sns.distplot(df_customer['Total day minutes'],bins=20)   
ax = f.add_subplot(gs[0, 1])
sns.distplot(df_customer['Total day calls'],bins=20)
ax = f.add_subplot(gs[0, 2])
sns.distplot(df_customer['Total day charge'],bins=20)
ax = f.add_subplot(gs[1, 0])
sns.distplot(df_customer['Total eve minutes'],bins=20)
ax = f.add_subplot(gs[1, 1])
sns.distplot(df_customer['Total eve calls'],bins=20)
ax = f.add_subplot(gs[1, 2])
sns.distplot(df_customer['Total eve charge'],bins=20)
f.tight_layout()
#Tât cả các plot đều tuân theo dạng chuẩn trong đó đỉnh nằm ở giữa đồ thị

f = plt.figure(figsize=(8,8))
gs = f.add_gridspec(2, 3)
ax = f.add_subplot(gs[0, 0])
sns.distplot(df_customer['Total night minutes'],bins=20)
ax = f.add_subplot(gs[0, 1])
sns.distplot(df_customer['Total night calls'],bins=20)
ax = f.add_subplot(gs[0, 2])
sns.distplot(df_customer['Total night charge'],bins=20)
ax = f.add_subplot(gs[1, 0])
sns.distplot(df_customer['Total intl minutes'],bins=20)
ax = f.add_subplot(gs[1, 1])
sns.distplot(df_customer['Total intl calls'],bins=20)
ax = f.add_subplot(gs[1, 2])
sns.distplot(df_customer['Total intl charge'],bins=20)
f.tight_layout()
#Tất cả đều dạng chuẩn và có đỉnh ở giữa đồ thị. Trừ total_intl call bị lệch phải

sns.catplot(x='State',kind='count',data=df_customer,height=8).set_xticklabels(rotation=60)
#WV là state có số lần xuất hiện nhiều nhấ

sns.catplot(x='International plan',kind='count',data=df_customer)
#Đa số có international plan là No với khoảng 3000

sns.catplot(x='Voice mail plan',kind='count',data=df_customer)
#Đa số là No với số lượng gần 2500

sns.catplot(x='Customer service calls',kind='count',data=df_customer)
#Số lượng trông giống phân phối chuẩn và có sốlượng nhiều nhất ở 1 với khoảng 1200

sns.catplot(x='Churn',kind='count',data=df_customer)
#False chiếm đa số với hơn 2500

-Phần văn bản
/*Chọn 2, 3, hay 4 tiêu chí bạn nghi ngờ có mối quan hệ với nhau mật thiết và biểu diễn chúng lên một hình EDA sau
đó cho nhận xét về mối quan hệ (Mỗi dữ liệu CustomerChurn hay BigMartSales cho 3 TH này)
BigMart Sales*/

sns.catplot(x='Outlet_Location_Type',y='Item_Outlet_Sales',hue='Outlet_Size',kind='box',data=df_mart)
#Có vẽ như Tier 2 chỉ có Outlet_Size ở dạng Small, Tier 1 không có Outlet có kích cỡ lớn và tier 3 không có outlet cỡ n

sns.catplot(x='Item_Type',y='Item_Outlet_Sales',kind='box',data=df_mart).set_xticklabels(rotation=90)
#Có vẻ như Hard Drink có lượng Item_Outlet_Sales thấp hơn cả. Trong khi đó với Item_type là Seafood thì người ta chi tiêu nhiều tiền hơn cho Item_Outlet

sns.catplot(x='Item_Type',y='Item_Weight',kind='box',data=df_mart).set_xticklabels(rotation=90)
#Những khách hàng mua household, Dairy, Health and Hygiene, Seafood và Others có xu hướng có cân nặng Item cao hơn những hàng mua Item type khác.

-Phần văn bản
/*
Customer Churn*/

sns.catplot(data=df_customer,x='International plan',y='Churn',hue='Voice mail plan',kind='bar')
#Với international plan No thì số lượng Voice mail plan với Churn đều thấp. Trong khi đó, tí lệ Voice mail plan và Chu
#Yes hay no thì đều đa số nằm ở International plan là Yes.

sns.catplot(data=df_customer,x='Churn',y='Customer service calls',kind='box')
#Với những khách hàng có Churn là True thì trung bình số cuộc gọi dịch vụ mà họ thực hiện
#Nhiều hơn rất nhiều so với khách hàng không Churn.

sns.catplot(data=df_customer,x='State',y='Total day minutes',kind='box',height=8).set_xticklabels(rotation=60)
#Các bang IN, PA, MD, ME có tổng số phút gọi ban ngày trung bình là cao hơn các bang còn lại.

sns.catplot(data=df_customer,x='State',y='Churn',kind='bar',height=8).set_xticklabels(rotation=60)
#Bang NJ, TX, CA, MD là các bang có tỉ lệ Churn cao nhất.
```
